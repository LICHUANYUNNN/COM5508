```python
import numpy as np
import pandas as pd
import warnings; warnings.filterwarnings('ignore')

# Read data from file
df = pd.read_csv('AppleStore.csv', encoding = 'utf8',index_col=0)
```

```python
df.head(5)
```

# Data cleaning

```python
# iOS
# 移除 评分为0的项目 空值 重复项
df = df.dropna()
df = df.drop_duplicates()
df = df[df['user_rating']!=0]
```

```python
df[['track_name','size_bytes','price','rating_count_tot','user_rating','sup_devices.num','ipadSc_urls.num','lang.num','prime_genre']]
```

```python
df[df['prime_genre']=='Catalogs']
```

# Variable coding

```python
# select numerical features
features = ['rating_count_tot','price','size_bytes','sup_devices.num','ipadSc_urls.num','lang.num'] #
```

```python
# select categorical features
category_list = df['prime_genre'].unique().tolist() 
category_list = ['cat_' + word for word in category_list]
df = pd.concat([df, pd.get_dummies(df['prime_genre'], prefix='cat',drop_first=True)], axis=1)
del category_list[-3]
features.extend(category_list)
```

# Linear Regression

```python
X = df[features]
y = df['user_rating']
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()

model.fit(X,y)
print('R2 using X:', model.score(X,y))
print('intercept:', model.intercept_)
print('coefficients:', model.coef_)
print('---------------------------------------------------------------------')

for i in features:
    df[i + '_log'] = np.log(df[i] + 1)
X_log = df[[i + '_log' for i in features]]

model.fit(X_log, y)
print('R2 using X_log: ', model.score(X_log, y))
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_log)
X_log_standardized = pd.DataFrame(scaler.transform(X_log), columns = X_log.columns)

model.fit(X_log_standardized, y)
print('R2 using X_log_standardized: ', model.score(X_log_standardized, y))
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```

```python
import statsmodels.api as sm
# Book为0
# Fit linear regression model using statsmodel package
X2 = sm.add_constant(X_log_standardized)
y = y.values.reshape(-1,1) # transfer y from pandas.series to 2D array

model = sm.OLS(y,X2).fit() 
model.summary()
```
